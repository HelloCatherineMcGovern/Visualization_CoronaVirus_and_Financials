<strong>Regarding the following pre-processing methods</strong>

Considerable cleanup was necessary for the dataset first dataset we employed (DXYOverall.csv). Some columns contained foreign characters and others simply contained URLs. Additionally there were null values everywhere. A considerable challenge lay in the fact that most of our data was contained within a single column (foreignData) in the form of a dictionary in each row of the column and about a dozen rows in the number of keys in the dictionary suddenly expanded. Thus a single data frame was created that included only the “foreignData” column as well as the date/time stamp. 
Additionally data was uploaded multiple times a day. For an overall view of each day’s activities, we chose to only display the maximum numbers. A loop was used to run through each date then using that date to create a max for the dataset then selecting the highest quantities for each day (Figure 1).

For the data on US cases, we opted for a different dataset. This data was organized with the dates as columns, instead of rows like the previous version and required different cleanup tactics. It was somewhat easier to pull in the dates into a plot when they were all listed in a single column.
To create the number of cities the virus had reached in a single day in the U.S. the dataset was converted into a data frame with only the rows and columns that were needed. Then the number of values that were greater than zero on that particular date were added together. In summary once the number of cases in that city grew beyond zero, the city was counted as a location that the virus had spread to.

#### <font color=Blue>-- We will update it to a scatter plot so it stands in better contrast to the Covid line overlaid. we are also able to program additional variables to visualize the size for the daily trade volume, and the color change between up and down days.</font>

### Results
The spread of the virus from the global to the local level all spread in similar fashion. The effects on markets were mixed from week to week. One can presume the markets rallied around specific economic news or trends descended based on negative perceptions about the future.

These charts show strong anecdotal correlation between the virus spread with individual company stock. Those online retailers appear almsot to be helped by the quarantine, while clearly brick-and-mortar retail and recreational companies are hit very hard.

Numerical spread and geographic spread appear to be correlated. 

Finally, we did not expect the data to show such strong evidence that "shelter-in-place" orders were effective and necessary. The final graph "Texas Cases and Daily increase" highlights that very fact by placing a marker on the day the shelter order was given by the Texas Governor, and the spike in cases precisely one incubation period after. Had the order not been given, all the bars to the right of the spike may have been yellow as well. It is enough to generate apprehension by government talks of an early return to work.
